# Business Understanding
#### Business Question
- customer mana saja yang punya potensi besar untuk membeli produk berdasarkan aktivitasnya di website?
#### Hypothesis
- manager perusahaan ingin mengetahui customer mana saja yang punya potensi besar untuk membeli produk berdasarkan aktivitasnya di website, sehingga perusahaan dapat menghubungi customer tersebut lebih cepat dengan treatment yang lebih tepat
#### Expected Output
- mengetahui customer yang punya potensi besar untuk membeli produk berdasarkan aktivitasnya di website
#### Experimentation Plan
- mengukur berapa persen potensi per customer untuk membeli produk

# Data Understanding
- Prospect ID : A unique ID with which the customer is identified.
<br>`ID unik yang digunakan untuk mengidentifikasi pelanggan.`
- Lead Number : A lead number assigned to each lead procured.
<br>`Nomor prospek yang ditetapkan untuk setiap prospek yang dibeli.`
- Lead Origin : The origin identifier with which the customer was identified to be a lead. Includes API, Landing Page Submission, etc.
<br>`Pengidentifikasi asal yang digunakan pelanggan untuk diidentifikasi sebagai prospek. Termasuk API, Pengajuan Halaman Arahan, dll.`
- Lead Source : The source of the lead. Includes Google, Organic Search, Olark Chat, etc.
<br>`Sumber timah. Termasuk Google, Pencarian Organik, Obrolan Olark, dll.`
- Do Not Email : An indicator variable selected by the customer wherein they select whether of not they want to be emailed about the course or not.
<br>`Variabel indikator yang dipilih oleh pelanggan di mana mereka memilih apakah mereka ingin diemail tentang kursus atau tidak.`
- Do Not Call : An indicator variable selected by the customer wherein they select whether of not they want to be called about the course or not.
<br>`Variabel indikator yang dipilih oleh pelanggan dimana mereka memilih apakah mereka ingin dipanggil tentang kursus atau tidak.`
- Converted : The target variable. Indicates whether a lead has been successfully converted or not.
<br>`Variabel sasaran. Menunjukkan apakah prospek berhasil dikonversi atau tidak.`
- TotalVisits : The total number of visits made by the customer on the website.
<br>`Jumlah total kunjungan yang dilakukan oleh pelanggan di situs web.`
- Total Time Spent on Website : The total time spent by the customer on the website.
<br>`Total waktu yang dihabiskan oleh pelanggan di situs web.`
- Page Views Per Visit : Average number of pages on the website viewed during the visits.
<br>`Jumlah rata-rata halaman di situs web yang dilihat selama kunjungan.`
- Last Activity : Last activity performed by the customer. Includes Email Opened, Olark Chat Conversation, etc.
<br>`Aktivitas terakhir yang dilakukan oleh pelanggan. Termasuk Email Dibuka, Percakapan Obrolan Olark, dll.`
- Country : The country of the customer.
<br>`Negara pelanggan.`
- Specialization : The industry domain in which the customer worked before. Includes the level 'Select Specialization' which means the customer had not selected this option while filling the form.
<br>`Domain industri tempat pelanggan bekerja sebelumnya. Termasuk level 'Pilih Spesialisasi' yang berarti pelanggan belum memilih opsi ini saat mengisi formulir.`
- How did you hear about Madugital : The source from which the customer heard about Madugital.
